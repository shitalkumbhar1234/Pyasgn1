```python
#6. Write a function to demonstrate the behavior of mutable and immutable arguments.
# Function to demonstrate immutable argument
x=[2,3,4,5]
def modify_immutable(x):
    print("Inside function before change:", x)
    x = x + 10
    print("Inside function after change:", x)

# Function to demonstrate mutable argument
def modify_mutable(lst):
    print("Inside function before change:", lst)
    lst.append(100)
    print("Inside function after change:", lst)
```

```python
#8. Write a program to demonstrate the use of logical operators.
# Taking user input
num = int(input("Enter a number: "))

# Using AND operator
if num > 0 and num % 2 == 0:
    print("The number is positive AND even")

# Using OR operator
if num < 0 or num % 2 == 1:
    print("The number is negative OR odd")

# Using NOT operator
if not (num == 0):
    print("The number is NOT zero")
```

```python
#9. Write a Python program to convert user input from string to integer, float, and boolean types.
user_input = input("Enter a value: ")

# Converting to different data types
int_value = int(user_input)
float_value = float(user_input)
bool_value = bool(user_input)

# Displaying results
print("Original value (string):", user_input, type(user_input))
print("Integer value:", int_value, type(int_value))
print("Float value:", float_value, type(float_value))
print("Boolean value:", bool_value, type(bool_value))
```

```python
#10. Write code to demonstrate type casting with list elements
# List of integers
num_list = [1, 2, 3, 4]

# Convert to float
float_list = [float(i) for i in num_list]

# Convert to string
string_list = [str(i) for i in num_list]

print("Original list:", num_list)
print("Float list:", float_list)
print("String list:", string_list)
```

```python
#13. Write a Python program to find the sum of all even numbers between 1 and 50.
total = 0

# Loop from 1 to 50
for i in range(1, 51):
    if i % 2 == 0:      # Check if number is even
        total += i      # Add even number to total

print("Sum of even numbers between 1 and 50 is:", total)
```

```python
#14. Write a program to reverse a string using a while loop.
text = input("Enter a string: ")
print(text)

# Initialize variables
reversed_text = ""
index = len(text) - 1   # Start from last character

# Using while loop
while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string is:", reversed_text)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop
num = int(input("Enter a number: "))

# Initialize variables
factorial = 1
i = 1

# Check for negative number
if num < 0:
    print("Factorial does not exist for negative numbers")
else:
    while i <= num:
        factorial *= i
        i += 1

    print("Factorial of", num, "is:", factorial)
```
