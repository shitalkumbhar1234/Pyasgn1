3. What are pre-defined keywords in Python

   -Pre-defined keywords in Python are reserved words that have a special meaning in the language.

You cannot use them as variable names, function names, or identifiers because Python already uses them for specific purposes.

They are part of Python’s core syntax.
Some commonly used keywords are:

if,else,elif,for,while,break,continue,def,return,class,try,except,import,True,False,None,and,or,not,pass,lambda,with,as,yield,global,nonlocal,assert

4. Can keywords be used as variable names

No, keywords cannot be used as variable names in Python.

Keywords are reserved words that have special meaning in Python’s syntax.

 Since Python already uses them to perform specific functions, they cannot be used as identifiers (variable names, function names, class names, etc.).

11. What is the purpose of conditional statements in Python

Conditional statements are used to make decisions in a program.

They allow the program to:
Execute certain code only if a condition is True
Skip or choose different code if the condition is False.

The purpose of conditional statements in Python is to control the flow of a program by executing different blocks of code based on whether a condition is True or False.

12. How does the elif statement work
It is used to check multiple conditions one by one after an if statement.
Python checks the if condition first.

If if is False → it checks the first elif.

If that is False → it checks the next elif.
When one condition becomes True → that block runs.

After that → Python stops checking further conditions.

If none are True → else block runs (if present).

🔹 Syntax
if condition1:

    # block 1
elif condition2:

    # block 2

elif condition3:

    # block 3
else:

    # block 4

🔹 Example
marks = 75

if marks >= 90:
    print("Grade A")

elif marks >= 70:

    print("Grade B")
elif marks >= 50:

    print("Grade C")
else:

    print("Fail")
Output:
Grade B
✔ Because:
marks >= 90 → False
marks >= 70 → True
So it prints Grade B and stops.
It does NOT check the next conditions after that.

13. What is the difference between for and while loops

**for loop** Used when Number of iterations is known.
 Based on	Sequence (list, range, string, etc.)
 Initialisation is automatic
 Risk of infinite loop is low

 **while loop**  Number of iterations is unknown
 Based on conditions
 Initialisation is manual
 Risk of infinite loop is high(if condition is not updated)
